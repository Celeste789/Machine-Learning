```python
def h_tita_multiple(x_i, tita):
    #is not the whole vector x, is the ith row value
    return np.dot(tita, x_i)
```

```python
def sum_gradient_descent_tita0_multiple(x, y, tita):
    aux = 0
    for i in range(0, len(y)):
        aux += (h_tita_multiple(x[i], tita) - y[i])
    return aux


def sum_cost_multiple(x, y, tita):
    aux = 0
    for i in range(0, len(y)):
        aux += (h_tita_multiple(x.iloc[i], tita)- y[i]) ** 2
    return aux


def sum_gradient_descent_tita_i(x, y, tita):
    aux = 0
    for i in range(0, len(x)):
        for j in range (0, len(tita)):
            aux += (h_tita_multiple(x.iloc[i], tita) - y[i]) * x.iloc[i, j]
    return aux

def gradient_descent_tita(tita, alpha, n_iterations, x, y):
    res = [tita]
    for i in range(0, n_iterations):
        for j in range(0, len(df2.columns) - 1):
            tita[j] = tita[j] - alpha / len(y) * sum_gradient_descent_tita_i(x, y, tita)
            res.append(tita.copy())
    return res

def cost(tita, x, y):
    costs = 1/2* len(y) * sum_cost_multiple(x, y, tita)
    return costs
```
